```python
from red_gym_env import PokeRedEnv
import numpy as np

# Initialize the environment
env = PokeRedEnv(
    gb_path="Pokemon Red.gb",  # Replace with the path to your Pokemon Red ROM
    init_state="init.state",  # Replace with the path to your initial state file
    headless=False  # Set to True if you don't want to see the game window
)

# Reset the environment to start a new game
observation, info = env.reset()

# Main game loop
for step in range(2):  # You can adjust the number of steps
    # Choose a random action
    action = env.action_space.sample()
    
    # Take a step in the environment
    observation, reward, terminated, truncated, info = env.step(action)
    
    # Print some information
    print(f"Step: {step}, Reward: {reward}, Action: {action}")
    
    # Check if the episode is done
    if terminated or truncated:
        print("Episode finished")
        break

# Close the environment
env.close()
```
